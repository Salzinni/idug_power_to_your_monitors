## Rolling mean in Pandas for Time Series Analysis
On this short example, we will use a CSV file containing around 150 observations for the number of connections an application has made to DDF daily, and we will quickly caculate a rolling mean with a window of 7 days, and then display the actual values and the mean trend in a Plotly chart.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline
sns.set()

df = pd.read_csv('App1.csv')

df['SMA_7'] = df.iloc[:,1].rolling(window=7).mean()

import plotly.graph_objects as go
fig = go.Figure()

fig.add_trace(go.Scatter(
                x = df.ds,
                y = df.y,
                name = 'Observations',
                line_color = 'deepskyblue',
                opacity = 0.8))

fig.add_trace(go.Scatter(
                x = df.ds,
                y = df['SMA_7'],
                name = 'MA 1 week',
                line_color = 'dimgray',
                opacity = 0.8))

fig.update_layout(title_text="App 1 connections made to DDF",
                   xaxis_rangeslider_visible=True)
fig.show()
```
